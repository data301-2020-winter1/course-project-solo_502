```python
import pandas as pd
import numpy as np
from functools import reduce



def load_and_process(path_to_cvs_file):
    
    # load data
    dataframe = pd.read_csv(path_to_cvs_file)


    # weather data by month
    df_monthly = dataframe.iloc[:,83:251]

    trying_months = []
    for i in range(14):
        trying_months.append(df_monthly.iloc[:,i*12:(i+1)*12].T.melt())
    
    trying_months2 = pd.concat(trying_months, axis = 1)
    trying_months3 = trying_months2.drop( columns = 'variable')
        
    
    # separating monthly variables, Transposing and melting puts all 12 monthly values in order and then in succession of years. 
    df_tmax = df_monthly.iloc[:,0:12].T.melt()
    df_tmin = df_monthly.iloc[:,12:24].T.melt()
    df_tave = df_monthly.iloc[:,24:36].T.melt()
    df_prec = df_monthly.iloc[:,36:48].T.melt()
    df_rad = df_monthly.iloc[:,48:60].T.melt()
    df_freeze = df_monthly.iloc[:,60:72].T.melt()
    df_above5 = df_monthly.iloc[:,72:84].T.melt()
    df_below18 = df_monthly.iloc[:,84:96].T.melt()
    df_above18 = df_monthly.iloc[:,96:108].T.melt()
    df_nffd = df_monthly.iloc[:,108:120].T.melt()
    df_pas = df_monthly.iloc[:,120:132].T.melt()
    df_eref = df_monthly.iloc[:,132:144].T.melt()
    df_cmd = df_monthly.iloc[:,144:156].T.melt()
    df_rh = df_monthly.iloc[:,156:168].T.melt()

    # Creating fitting dataframe with years as variable to fit monthly data 
    period = np.arange(1901, 2020, 1).repeat(12)
    period2 = np.vstack(period)
    years = pd.DataFrame(data=period2)

    
    # Creating a list for the dataframe to merge them. Renaming columns. Unsure why some monthly variables measure yearly numbers though
    monthly_dataframes = [years, df_tmax['value'], df_tmin['value'], df_tave['value'], df_prec['value'], 
                          df_rad['value'], df_freeze['value'], df_above5['value'], df_below18['value'], 
                          df_above18['value'], df_nffd['value'], df_pas['value'], df_eref['value'], 
                          df_cmd['value'], df_rh['value']]

    
    
    # Fitting monthly variables into a single dataframe
    df_months = pd.concat(monthly_dataframes, join='outer', axis = 1)
    df_months.columns = ['Period','Max Temp', 'Min Temp', 'Ave Temp', 'Precipitation', 'Radiation', 'Days < 0°C', 'Days < 5°C', 
                'Days < 18°C', 'Days > 18°C', 'Frost Free Days', 'Snowfall', 'Evaporation', 
                'Moisture Deficit', 'Relative Humidity']

    
    print(df_months)
    print(trying_months3)
    
load_and_process(r'C:\Users\Len\Documents\GitHub\solo_502\course-project-solo_502\data\raw\ClimateData_UBCO.csv')
```
